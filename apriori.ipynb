```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
data=pd.read_csv("basket_analysis.csv")
```

```python
data
```

```python
df=data.drop('Unnamed: 0', axis=1)
```

```python
df
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
frequent_skillsets=apriori(df, min_support=0.17, use_colnames=True)
frequent_skillsets
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
res=association_rules(frequent_skillsets, metric="confidence", min_threshold=0.47)
res=res[['antecedents','consequents','confidence']]
res = res.sort_values(by="confidence", ascending=False)
res
```

```python
import re

for index, row in res.iterrows():
    antecedents = ', '.join([re.sub(r'skill_', '', item).replace('_', ' ').title() for item in row['antecedents']])
    consequents = ', '.join([re.sub(r'skill_', '', item).replace('_', ' ').title() for item in row['consequents']])
    confidence = round(row['confidence'], 3)
    print(f"{antecedents} -> {consequents}. Confidence = {confidence}")
```
